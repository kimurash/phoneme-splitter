```python
import librosa
import soundfile as sf

import wave
import matplotlib.pyplot as plt
```

音声ファイルを読み込んで波形を表示してみる．

```python
SAMPLE_WAV = '../wav/C0002.wav'

y, sr = librosa.load(SAMPLE_WAV, sr=16000, mono=True)
plt.plot(y);
```

soundfileで量子化bit数を変更して結果を確認する．

```python
sf.write('result.wav', y, sr, subtype='PCM_16')
```

```python
wv = wave.open('./result.wav', 'rb')

print(wv.getnchannels())
print(wv.getsampwidth()) # 1サンプルのバイト数
print(wv.getframerate()) # サンプリング周波数
print(wv.getnframes())
```
